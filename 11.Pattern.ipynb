```python
n = int(input("Enter a number:"))

i = 1
while i <= 2*n - 1:

    k = i if i <= (2*n - 1)//2 else 2*n - i

    # spaces
    j = 1
    while j <= n - k:
        print(" ", end="")
        j += 1

    # stars
    p = 1
    while p <= 2*k - 1:
        print("*", end="")
        p += 1

    print()
    i += 1
```

```python
n = 5

i = 1
while i <= n:

    # space ka  liye loop hai j
    j = 1
    while j <= n - i:
        print(" ", end="")
        j += 1

    # numbers print kar rhe increasing and decreasing order maii
    num = i
    k = 1
    while k <= 2*i - 1:

        print(num, end="")

        if k < i:
            num += 1
        else:
            num -= 1

        k += 1

    print()
    i += 1
```
