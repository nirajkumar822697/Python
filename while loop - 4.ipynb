# ***`TO-DO LIST`***

- Add Task
- View
- Delete
- change
- exit

```python
todo = []
```

```python
while True:
    choice = int(input("1.Add Task\n2.View Task\n3.Delete Task\n4.Change Task\n5.Exit Task\n"))

    if choice == 1:
        task = input("Enter Your Task")
        todo.append(task)  # add item in the end of list
        print(f"{task} has been added...")

    elif choice == 2:
        if todo:
            i = 0
            while i < len(todo):
                print(f"Task{i} : {todo[i]}")
                i += 1

        else:
            print("Todo is empty....")

    elif choice == 3:
        if todo:
            task = input("Enter which task you want to remove ?")
            if task in todo:
                index_value = todo.index(task)
                delete_value = todo.pop(index_value)
                print(f"{delete_value} has been removed")

            else:
                print("Task is not available...")
            

        else:
            print("Todo is empty")


    elif choice == 4:
        if todo:
            task = input("Enter your task to modify....")
            if task in todo:
                idx = todo.index(task)
                new_task = input("Enter new task...")
                todo[idx] =new_task
                print("Task modified.....")
                

            else:
                print("Task is not avialable")

        else:
            print("Todo is empty")
        
            






    elif choice == 5:
        print("Exit from Todo list..")
        break
```

```python
Create empty list students

While True:

    Show menu

    Take user choice

    If add:
        Take name, roll, marks
        Store in dictionary
        Add to list

    If view:
        Show all students

    If delete:
        Remove by roll number

    If update:
        Change marks/name

    If exit:
        Break
```

```python
#### ls = []
student = {}


while True:
    choice = int(input("1.Add Student\n2.View Student\n3.Delete Student\n4.Exit Task\n"))
    # Add task
    if choice ==1:
        name = input("Enter the student name: ")
        roll_no = int(input("Enter your Roll number: "))
        marks = float(input("Enter your marks: "))
        student = {
            "name": name,
            "roll": roll_no,
            "marks": marks
        }
        ls.append(student)
        print("Student added!")
    
    # view student
    elif choice == 2:
        if not ls:
            print("List is empty")
        else:
            i = 0
            while i < len(ls):
                print(ls[i])
                i += 1
    
    # delete student
    elif choice == 3:
        if not ls:
            print("List is empty")

        else:
            roll_no = int(input("Enter roll to delete: "))

            for student in ls:

                if student["roll"] == roll_no:

                    ls.remove(student)
                    print("Student deleted")
                    break

            else:
                print("Student not found")
    # exit
    elif choice == 4:
        print("Exit from Student list..")
        break       
```
