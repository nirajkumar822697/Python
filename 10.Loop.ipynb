### 1.Python Program to Check Palindrome Number 

```python
num = int(input("Enter a number: "))
temp =num
rev=0
rem=0
while(temp>0):
    rem =temp%10
    rev = rev*10+rem
    temp =temp//10
if num ==rev:
    print(f"{num} is a palindrome number")
else:
    print(f" {num} is a not palindrome number")
```

```python
num = int(input("Enter a number: "))
temp =num
rem = 0
count  = 0
while(temp>0):
    rem = temp%10
    count+=1
    temp = temp//10

sum =0
temp = num
while(temp>0):
    rem = temp % 10
    sum+= rem**count
    temp = temp//10
if sum == num:
    print("This is armstorng number")
else:
    print("This is  not a armstorng number")
    
```

### 14.Convert a string to title case (first letter of each word capital) using loop. 

```python
Str = input("Enter string: ")
str2 =""
i=0
while(i<len(Str)):
    if i==0:
        str2+= Str[i].capitalize()
    elif Str[i-1] == " ":
        str2+= Str[i].capitalize()
    else:
        str2+=Str[i]
    i+=1
print(str2)
```
