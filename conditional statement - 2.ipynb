# Nested If
    - if inside a if

```python
ls = ["rohan" , "aman" , "rohan" , "meenakshi"]
ls
```

```python
if ls:  # >> True
    if ls.pop():  #>>> meenakshi
        print(ls)  # >>> ls
    else:
        print("bye")
else:
    print("Hello")
```

```python
if ls.sort():   ## > None
    print("hi")
    if ls.remove("rohan"):
        print("rohan")
    else:
        print("Goodbye")
else:
    print(ls)  # list
```

```python
16.	Write a Python program to input marks of five subjects Physics, Chemistry, Biology, Mathematics and
Computer. Calculate percentage and grade according to following: 
1.	Percentage >= 90% : Grade A
2.	Percentage >= 80% : Grade B 
3.	Percentage >= 70% : Grade C 
4.	Percentage >= 60% : Grade D 
5.	Percentage >= 40% : Grade E 
6.	Percentage < 40% : Grade F
```

```python
phy = float(input("Enter marks of physics"))
chem = float(input("Enter marks of chemistry"))
bio = float(input("Enter marks of biology"))
maths = float(input("Enter marks of maths")) 
comp = float(input("Enter marks of computer"))

total_marks = phy + chem + bio + maths + comp
percentage = (total_marks / 500) * 100
if percentage >= 90:
    print("Grade A")
elif percentage >= 80:
    print("Grade B")
elif percentage >= 70:
    print("Grade C")
elif percentage >= 60:
    print("Grade D")
elif percentage >= 40:
    print("Grade E")
else:
    print("Grade F")
```

***`Write a Python program to input basic salary of an employee and calculate its Gross salary according to following:`***
- i.	Basic Salary <= 10000 : HRA = 20%, DA = 80% 
- ii.	Basic Salary <= 20000 : HRA = 25%, DA = 90%
- iii.  Basic Salary > 20000 : HRA = 30%, DA = 95%


```python
salary = int(input("Enter your salary"))
if salary <= 10000:
    hra = salary * 0.20
    da = salary * 0.80
    print(f"HRA IS {hra}")
    print(f"DA is {da}")

elif salary <= 20000:
    hra = salary * 0.25
    da = salary * 0.90
    print(f"HRA IS {hra}")
    print(f"DA is {da}")

elif salary > 20000:
    hra = salary * 0.30
    da = salary * 0.95
    print(f"HRA IS {hra}")
    print(f"DA is {da}")
```
