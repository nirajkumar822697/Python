## **Odd/Even Algorithm**

1. Start the program
2. Take a input from the user
3. if number completely divide by 2 print Even
4. else print odd
5. End the program 

1. Start the program
2. Take a number input from the user
3. if number is equal to 0 print Neutral
4. if number is greater than 0 print positive
5. else print negative
6. End the program


## *Pseudo code*
    - looks like code but its not a code

- start
- input a number
- if number % 2 == 0
    - print even
- else
    - print odd 



```python
#print prime or not prime number
num = int(input("Enter a number: "))

if num <= 1:
    print("Not a prime number")
else:
    for i in range(2, num // 2 + 1):
        if num % i == 0:
            print("Not a prime number")
            break
    else:
        print("Prime number")
```
