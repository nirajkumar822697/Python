#### 1.Program to find the sum of digits of a number until the sum is reduced to 1 digit 

```python
num = int(input("Enter a number: "))
rem=0
sum=0
while num > 0:
    rem = num%10
    sum += rem
    num //=10

print(sum)
```

#### 2. Program to find LCM and HCF of two numbers 

```python
# HCF
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

hcf = 1


for i in range(1, a+1):
    if a % i == 0 and b % i == 0:
        hcf = i

print(f"HCF  of {a} and {b} is", hcf)
```

```python
# LCM
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

Lcm = 2

while True:
    if Lcm % a == 0 and Lcm % b == 0:
        break
    Lcm += 1  # next number check karo

print(f"LCM of {a} and {b} is:", Lcm)
```

#### 3.Write a program to accept any number n and print the cube of all numbers from 1 to n which are divisible by 3. Rewrite the program using a continue statement. 

```python
num = int(input("Enter a number"))
i=1
while i<= num:
    if i%3==0:
        print(i**3)
    i+=1
```

```python
# using a continue statement
num = int(input("Enter a number"))
i=1
while i<= num:
    if i%3!=0:
        i+=1
        continue
    else:
        print(i**3)
        i+=1
  
```

#### 3.Write a program to read roll number and marks of 10 students in 3 subjects. The valid range for roll number is 1000-9999, if the roll number entered is not in this range, the user should be asked to enter again. Calculate total marks of only those students who get more than or equal to 40 marks in each subject. Count the number of students whose total is more than 200. Print the roll number of the students who gets the highest total. 

```python
students = []

i = 1
while i <= 2:
    
    roll_no = int(input(f"Enter roll no of {i} student: "))
    
    while roll_no < 1000 or roll_no > 9999:
        print("Invalid roll number! Enter again.")
        roll_no = int(input(f"Enter roll no of {i} student: "))
    
    marks = []
    
    for j in range(1, 4):
        m = float(input(f"Enter marks in subject {j}: "))
        marks.append(m)
    
    students.append([roll_no, marks])
    
    i += 1

print(students)
```

```python
# total marks who score greater than 40 in each subject
total_marks = 0

for student in students:
    marks = student[1]
    
    if marks[0] >= 40 and marks[1] >= 40 and marks[2] >= 40:
        total_marks += marks[0] + marks[1] + marks[2]

print("Total marks of passed students:", total_marks)
```

```python
#Count the number of students whose total is more than 200. 
count = 0

for student in students:
    marks = student[1]
    total = marks[0] + marks[1] + marks[2]
    
    if marks[0] >= 40 and marks[1] >= 40 and marks[2] >= 40:
        if total > 200:
            count += 1

print("Number of students with total > 200:", count)
```

```python
highest_total = 0
roll_no_topper= 0

for student in students:
    roll = student[0]
    marks = student[1]
    
    total = marks[0] + marks[1] + marks[2]
    
    if marks[0] >= 40 and marks[1] >= 40 and marks[2] >= 40:
        if total > highest_total:
            highest_total = total
            roll_no_topper = roll

print("Roll number of highest total student:",roll_no_topper)
```
